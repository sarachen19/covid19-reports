```python
import pandas as pd
import plotly.graph_objects as go
groupByCountry = worldData.groupby(worldData['name_en'])
worldFig = go.Figure()
for country, group in groupByCountry:
    tempDf = pd.DataFrame(data = group)
    tempDf['cumsum'] = tempDf['cases'].cumsum()
    worldFig.add_trace(go.Scatter(x=group['date'], y = group['cumsum'], mode='lines', name = country))
    #print(country)
    #print(group)
worldFig.show()
```

```python
import plotly.express as px    
def f_top(df):
    return df.sort_values(by = ["cumsum"], ascending = False, ignore_index = True)[['name_en','date',"cumsum"]].iloc[:10]
worldData['cumsum'] = worldData.groupby(worldData['name_en'])['cases'].cumsum()
#worldAnimationDf['date'] = worldAnimationDf['date'].dt.strftime('%Y-%m-%d')
worldDataGroup = worldData.groupby(['date']).apply(f_top)
worldDataGroup = worldDataGroup.loc[worldDataGroup['date'] > '2020-02-11']
#print(worldDataGroup.loc[worldDataGroup['date'] == '2020-03-11'])

worldDataAnimation = px.bar(worldDataGroup, x =worldDataGroup['cumsum'], y = worldDataGroup['name_en'],  animation_frame='date', animation_group="name_en", color = "name_en")
#worldDataAnimation.update_xaxes(categoryorder = 'category descending')
worldDataAnimation.show()
```

## world data-date slidebar and animation

```python
#set cumsum of each country
worldDataUpdate = worldData[['date', 'name_en', 'deaths', 'cases']].copy()
worldDataUpdate['cumsum'] = worldDataUpdate.groupby(worldDataUpdate['name_en'])['cases'].cumsum()
worldDataUpdate['date'] = pd.to_datetime(worldDataUpdate['date'], format = '%Y-%m-%d')
#get unique countries and color list
countries = worldDataUpdate['name_en'].unique()
countries_count = len(countries)
colors1 = px.colors.qualitative.Pastel
colors2 = px.colors.qualitative.Vivid
colors3 = px.colors.qualitative.Dark2
color_list = (colors1 + colors2 + colors3)*9
color_list = color_list[:countries_count]

#set color for each country
for i in range(countries_count):
        worldDataUpdate.loc[worldDataUpdate['name_en'] == countries[i], 'color'] = color_list[i] 
        
#save update into csv
worldDataUpdate.to_csv("worldData_update.csv")
```

```python
import numpy as np
import plotly.graph_objects as go
import datetime

date_country_cumsum_dict = {} #key: date, value: dataframe of date,name,cumsum
dates = np.sort(pd.unique(worldDataUpdate['date']))
for date in dates:
    countries_cumsum_df = worldDataUpdate.loc[worldDataUpdate['date'] == date]
    countries_cumsum_df = countries_cumsum_df.nlargest(20, 'cumsum').sort_values(by = ['cumsum'], ascending = True)
    date_country_cumsum_dict[date] = countries_cumsum_df
#print(date_country_cumsum_dict)

# Add traces, one for each slider step
worldFig = go.Figure()
for date in date_country_cumsum_dict:
    tempDf = date_country_cumsum_dict[date]
    worldFig.add_trace(
        go.Bar(
            name = "",
            text =tempDf['cumsum'],
            textfont = dict(size = 14, family = 'Droid Sans Mono'),
            textposition = "inside",
            texttemplate = '%{text:.3s}',
            visible = False,
            marker = dict(color = tempDf['color']),
            x = tempDf['cumsum'],
            y = tempDf['name_en'],
            orientation='h',
        )
    )
worldFig.data[185].visible = True

# Create and add slider
figSize = len(worldFig.data)
steps = []
visibleIndex = 0
for date in date_country_cumsum_dict:
    tempDf = date_country_cumsum_dict[date]
    step = dict(
    method = 'update',
    args = [{"visible": [False] * figSize}], #set all traces to invisible
    label = str(pd.to_datetime(date).date())
    )
    step['args'][0]["visible"][visibleIndex] = True #set one day's trace to visible. visibleIndex0 <=> date0, visibleIndex1 <=> date1 ...
    visibleIndex += 1
    steps.append(step)
sliders = [dict(
    active = 185,
    steps = steps,
)]

#print(steps[visibleIndex - 2]["args"])
worldFig.update_layout(
    sliders = sliders,
    width=900,
    height=700,
    margin=dict(t=20, b=0, l=0, r=0),
    template="plotly_white",
)
worldFig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
import datetime

date_country_cumsum_dict = {} #key: date, value: dataframe of date,name,cumsum
dates = np.sort(pd.unique(worldDataUpdate['date']))
for date in dates:
    countries_cumsum_df = worldDataUpdate.loc[worldDataUpdate['date'] == date]
    countries_cumsum_df = countries_cumsum_df.nlargest(20, 'cumsum').sort_values(by = ['cumsum'], ascending = True)
    date_country_cumsum_dict[date] = countries_cumsum_df

frames = []
for date in date_country_cumsum_dict:
    tempDf = date_country_cumsum_dict[date]
    maxCumsum = tempDf['cumsum'].max()
    dateString = str(pd.to_datetime(date).date())
    frame = go.Frame(
        data = [go.Bar(
            name = "",
            text =tempDf['cumsum'],
            textfont = dict(size = 14, family = 'Droid Sans Mono'),
            textposition = "inside",
            texttemplate = '%{text:.3s}',
            marker = dict(color = tempDf['color']),
            x = date_country_cumsum_dict[date]['cumsum'],
            y = date_country_cumsum_dict[date]['name_en'],
            orientation='h',
        )],
         layout = go.Layout(
             xaxis = dict(range=[0,maxCumsum], autorange=False),
            # yaxis = dict(range=[0, 25], autorange=False),
             title = dict(text=('Confirmed numbers per Country: ' + dateString), font=dict(size=28),x=0.5,xanchor='center'),
         )
    )
    frames.append(frame)

worldFig = go.Figure( 
    data=[go.Bar(
            name = "",
            marker = dict(color = tempDf['color']),
            x = date_country_cumsum_dict[np.datetime64('2019-12-31T00:00:00.000000000')]['cumsum'],
            y = date_country_cumsum_dict[np.datetime64('2019-12-31T00:00:00.000000000')]['name_en'],
            orientation='h',
        )],
    layout=go.Layout(
        title=dict(text='Confirmed numbers per Country: 2019-12-31',font=dict(size=28),x=0.5,xanchor='center'),
        xaxis=dict(range=[0,4000000], autorange=False),
       # yaxis=dict(range=[0, 25], autorange=False),
        updatemenus=[dict(
            type="buttons",
            direction="left",
            x=-0.3,
            y=1,
            buttons=[dict(label="Play",
                          method="animate",
                          args=[None,
                          {"frame": {"duration": 500, "redraw": True},
                          "transition": {"duration":100,
                          "easing": "linear"}}]
            )]
        )],
     ),
    frames = frames,
)

worldFig.show()
```
